```python
# Matplotlib va NumPy kutubxonalarini import qilish
import matplotlib.pyplot as plt
import numpy as np

# Sinus funksiyasining grafiki uchun ma'lumotlar
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Scatter grafik uchun tasodifiy nuqtalar
x_scatter = np.random.rand(100) * 10
y_scatter = np.random.rand(100) * 10

# Grafiklarni chizish
plt.figure(figsize=(10, 5))

# Sinus funksiyasining grafiki
plt.subplot(1, 2, 1)
plt.plot(x, y, label='Sinus(x)')
plt.xlabel('X qiymatlari')
plt.ylabel('Y qiymatlari')
plt.title('Sinus Funksiyasining Grafiki')
plt.legend()

# Tasodifiy nuqtalar scatter grafiki
plt.subplot(1, 2, 2)
plt.scatter(x_scatter, y_scatter, label='Nuqtalar')
plt.xlabel('X qiymatlari')
plt.ylabel('Y qiymatlari')
plt.title('Tasodifiy Nuqtalar Scatter Grafiki')
plt.legend()

# Grafikni saqlash
plt.tight_layout()
plt.savefig('graphics.png')

# Grafikni ko'rsatish
plt.show()
```
